```python
import numpy as np                                                         #Import de la librería numpy
```

```python
# --- EXTRACCIÓN DE DATOS (EXTRACT) ---
```

```python
df_customer = pd.read_csv("customer_data.csv")    #Creación del primer data frame con el dataset "customer_data"
```

```python
df_sales = pd.read_csv("sales_data.csv")          #Creación del segundo data frame con el dataset "sales_data"
```

```python
df_customer                                                                  #Acceso a los datos del primer data frame
```

```python
df_sales                                                            #Acceso a los datos del segundo data frame
```

```python
df_customer.info()                                #Información general del data frame customer
```

```python
df_sales.info()                                #Información general del data frame customer
```

```python
# --- Verificación y gestión la unicidad de los id del cliente --- 
```

```python
conteo_duplicados = df_customer['customer_id'].value_counts()
```

```python
ids_duplicados = conteo_duplicados[conteo_duplicados > 1]
```

```python
print(f"Número total de IDs de clientes duplicados: {len(ids_duplicados)}")
```

```python
# --- Concatenación de los data frames ---
```

```python
df = pd.merge(df_customer,df_sales, on="customer_id", how="right")    #Unión de los dos data frames manteniendo todos los registros de sales
```

```python
df                                                                    #Acceso a los datos 
```

```python
# --- TRANSFORMACIÓN DE DATOS (TRANSFORM) ---
```

```python
# --- GESTION DE LOS VALORES NULOS -- 
# Debio al bajo porcentaje de valores nulos decidimos eliminar dichas filas
```

```python
filas_antes=len(df)                                             
print(f"La cantidad de filas ante de elimiar los nulos es:{filas_antes}")
```

```python
df_notnull = df.dropna(subset=["age"], inplace=False)           #Eliminacion de filas con age nulo sin modificar el df original
```

```python
filas_despues = len(df_notnull)
print(f"La cantidad de filas luego de elimiar los nulos es: {filas_despues}")
```

```python
df_notnull
```

```python
# --- ESTANDARIZACIÓN FORMATO DE FECHA ---
```

```python
df_notnull.loc[:,"invoice_date"] = pd.to_datetime (df_notnull["invoice_date"], errors = 'coerce', dayfirst=True)
```

```python
df_notnull["invoice_date"]
```

```python
df_notnull.info()
```

```python
df_notnull
```

```python
# --- OPERACIONES ---
```

```python
# Modo de pago más frecuente - Usamos .mode()[0]
```

```python
modo_pago_general = df_notnull["payment_method"].mode()[0]
```

```python
print(f"El modo de pago más frecuente es: {modo_pago_general}")
```

```python
# Modo de pago más frecuente por género
```

```python
modo_pago_generos = df_notnull.groupby('gender')["payment_method"].agg(lambda x: x.mode()[0])
```

```python
print(f"El modo de pago más frecuente por género es: {modo_pago_generos}")
```

```python
conteo_pago = df_notnull['payment_method'].value_counts()                    #Verificación 
```

```python
print(f"Conteo de todas las categorías:{conteo_pago}")
```

```python
conteo_genero = df_notnull.groupby('gender')['payment_method'].value_counts()
```

```python
print(f"Conteo de todas las categorías por géneros:{conteo_genero}")         #Verificación
```

```python
# Concatenación de Método de pago agrupado por género y edad
```

```python
df_pagos = pd.DataFrame(df_notnull.groupby(['gender','age'])['payment_method'].value_counts())
df_pagos.rename(columns={'count': 'Conteo_Transacciones'}, inplace=True)
```

```python
df_pagos
```

```python
# Métodos de pagos realizados por el rango etario de 25 a 35 años 
```

```python
rango_etario_df = df_notnull[(df_notnull['age']>=25) & (df_notnull['age']<=35)]
```

```python
pago_rango_etario = pd.DataFrame(rango_etario_df['payment_method'].value_counts())
pago_rango_etario.rename(columns={'count': 'Conteo_Transacciones'}, inplace=True) 
```

```python
pago_rango_etario
```

```python
# Métodos de pago más utilizados por las mujeres
```

```python
mujeres = df_notnull[df_notnull['gender']=="Female"]
```

```python
metodo_mujeres = mujeres['payment_method'].mode()[0]
```

```python
print(f"Método de pago más utilizado por mujeres es: {metodo_mujeres}")
```

```python
# Precios por categoría de productos 
```

```python
precios_categoria = df_notnull.groupby('category')['price'].mean().sort_values(ascending=False).round() 
```

```python
precios_categoria
```

```python
precios_por_categoria = df_notnull.groupby('category')['price'].agg(['mean', 'min', 'max', 'count']).reset_index().sort_values(by='mean', ascending=False).round() 
print("Precios por categoría de productos:")
print(precios_por_categoria)
```
